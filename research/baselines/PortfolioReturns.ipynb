```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
companies_baselines = pd.read_csv('./COMPANIES_BASELINES.csv')
backtests_returns = pd.read_csv('./BACKTESTS_BASELINES.csv')
alpha_returns = pd.read_csv('./ALPHAS.csv')
```

```python
backtests_returns['strategy_config_id'] = [i for i in range(1, 41)] * 50

time_periods = list(companies_baselines.columns)
time_periods.remove('ticker')

return_values = {}
return_values['strategy_config'] = []
for key in time_periods:
	return_values[key] = []

return_values['strategy_config'].append(0)
for time_period in time_periods:
	return_values[time_period].append(sum(companies_baselines[time_period]))

for strategy_config in range(1, 41):
	return_values['strategy_config'].append(strategy_config)
	strategy_config_returns = backtests_returns[backtests_returns['strategy_config_id'] == strategy_config]
	for time_period in time_periods:
		# print(strategy_config, time_period, sum(strategy_config_returns[time_period]))
		return_values[time_period].append(sum(strategy_config_returns[time_period]))

portfolio_returns = pd.DataFrame(columns=return_values.keys())
for key in return_values:
	portfolio_returns[key] = return_values[key]
portfolio_returns
```

```python
portfolio_returns.to_csv('./PORTFOLIO_RETURNS.csv')
```

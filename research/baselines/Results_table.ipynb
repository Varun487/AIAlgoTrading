# Tabulate Results of Strategy Backtests

### 1. Imports

```python
import pandas as pd
```

### 2. Analyze data

```python
company_baselines = pd.read_csv('./COMPANIES_BASELINES.csv')
backtest_baselines = pd.read_csv('./BACKTESTS_BASELINES.csv')
```

```python
company_baselines.head()
```

```python
backtest_baselines.head()
```

```python
LSTM_baselines = backtest_baselines[backtest_baselines['strategy_type_id'] == 2]
LSTM_baselines.head()
```

```python
BB_baselines = backtest_baselines[backtest_baselines['strategy_type_id'] == 1]
BB_baselines.head()
```

### 3. Calculations

```python
# Create data frame
df = pd.DataFrame(columns=["Strategy", "Profitable Time Periods", "Baseline Beating Time Periods", "Percentage Profitable Time Periods", "Percentage Baseline Beating Time Periods", "Total Time Periods"])
```

```python
# Add strategies
df["Strategy"] = ["LSTM", "BB", "Combined"]
```

```python
# Calculate profitable time periods

lstm_total = 0
# LSTM Strategy
lstm_profitable_timeperiods = 0
for col in LSTM_baselines.columns[4:]:
	# print(col)
	for baseline_return in LSTM_baselines[col]:
		lstm_total += 1
		if baseline_return > 0:
			# print(baseline_return)
			lstm_profitable_timeperiods += 1
# print(lstm_profitable_timeperiods, "/", lstm_total)

bb_total = 0
# BB Strategy
bb_profitable_timeperiods = 0
for col in BB_baselines.columns[4:]:
	# print(col)
	for baseline_return in BB_baselines[col]:
		bb_total += 1
		if baseline_return > 0:
			# print(baseline_return)
			bb_profitable_timeperiods += 1
# print(bb_profitable_timeperiods, "/", bb_total)

df["Profitable Time Periods"] = [lstm_profitable_timeperiods, bb_profitable_timeperiods, lstm_profitable_timeperiods + bb_profitable_timeperiods]
df["Total Time Periods"] = [69000, 69000, 138000]
df["Percentage Profitable Time Periods"] = [
	(lstm_profitable_timeperiods / lstm_total)*100,
	(bb_profitable_timeperiods / bb_total)*100,
	((lstm_profitable_timeperiods + bb_profitable_timeperiods) / (bb_total + lstm_total))*100,
]
```

```python
# Calculate baseline beating time periods
alphas = pd.read_csv("./ALPHAS.csv")

lstm_alphas = alphas[alphas['strategy_type_id'] == 2] 
bb_alphas = alphas[alphas['strategy_type_id'] == 1]

df["Baseline Beating Time Periods"] = [len(lstm_alphas), len(bb_alphas), len(lstm_alphas) + len(bb_alphas)]
df["Percentage Baseline Beating Time Periods"] = [
	(len(lstm_alphas) / 69000) * 100, 
	(len(bb_alphas) / 69000) * 100, 
	((len(lstm_alphas) + len(bb_alphas)) / 130000) * 100
]
```
